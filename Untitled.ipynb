```python
import requests
import pandas as pd
```

```python
url = "https://www.omscentral.com/_next/data/RbJpKU_7gp7Gm26pP9748/reviews/new.json"

response = requests.get(url)
data = response.json()
data['pageProps']['courses']
```

```python
url = "https://www.omscentral.com/_next/data/RbJpKU_7gp7Gm26pP9748/index.json"

response = requests.get(url)
data = response.json()
courses = data['pageProps']['courses']
```

```python
len([course for course in courses if course['reviewCount'] > 3])
```

```python
courses_df = pd.DataFrame(courses)
courses_df.columns
```

```python
courses_df[(courses_df['isFoundational'])].plot(kind="scatter", x="difficulty", y="rating", s="workload")
```

```python
courses_df[~(courses_df['isFoundational'])][['name', 'codes', 'tags', 'rating', 'difficulty', 'workload']]
```

```python
courses_df[(courses_df['isFoundational'])][['rating','difficulty', 'workload']].median(numeric_only=True)
```
